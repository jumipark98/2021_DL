## About the U-Net

- U-Net is a segmentation network based on fully convolutional network. The key idea of U-Net is applying the cropped feature map concatenation between the downsampling path and upsampling path.

- ref : https://lmb.informatik.uni-freiburg.de/people/ronneber/u-net/

![대체 텍스트](https://lmb.informatik.uni-freiburg.de/people/ronneber/u-net/u-net-architecture.png)



## About Semantic Boundaries Datasety

- Hariharan (2011) created the Semantic Boundaries Dataset(henceforth abbreviated as SBD) and the associated benchmark to evaluate the task of predicting semantic contours, as opposed to semantic segmentations. While semantic segmentation aims to predict the pixels that lie inside the object, SBD interested in predicting the pixels that lie on the boundary of the object.

- ref : http://home.bharathh.info/pubs/codes/SBD/download.html

![대체 텍스트](https://www2.eecs.berkeley.edu/Research/Projects/CS/vision/grouping/icons/semantic_contours.png)

### Dataset
The **SBD** currently contains annotations from **11355 images** taken from the **PASCAL VOC 2011 dataset**.These images were annotated on Amazon Mechanical Turk and the conflicts between the segmentations were resolved manually. For each image, They provide both category-level and instance-level segmentations and boundaries. The segmentations and boundaries provided are for the **20 object categories** in the **VOC 2011 challenge**.

````python
import torch
import torchvision
import numpy as np
import torchvision.transforms as transforms
from PIL import Image
import cv2
from albumentations import CLAHE,Compose
#TODO: change value with your code.
IMAGE_S = 224
BATCH_S = 8
clahe = CLAHE(p=0.8)

def transform(image, target):
    image_transform = transforms.Compose([
        transforms.ColorJitter(0.2,0.2,0.2,0),
        transforms.RandomGrayscale(p=0.1),
        transforms.Resize((IMAGE_S, IMAGE_S)),
        transforms.ToTensor(),
        transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224,0.225))
    ])

    target_transform =  transforms.Compose([
        transforms.Resize((IMAGE_S, IMAGE_S), interpolation=Image.NEAREST),
        np.asarray,
        torch.LongTensor
    ])
    image =cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
    data = {"image":(image)}
    aug = clahe(**data)
    png_img = aug["image"]
    png_img = Image.fromarray(png_img)

    return image_transform(png_img), target_transform(target)
def test_transform(image,target):
    image_transform = transforms.Compose([
        transforms.Resize((IMAGE_S, IMAGE_S)),
        transforms.ToTensor(),
        transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224,0.225))
    ])
    target_transform =  transforms.Compose([
        transforms.Resize((IMAGE_S, IMAGE_S), interpolation=Image.NEAREST),
        np.asarray,
        torch.LongTensor
    ])
    image =cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
    data = {"image":(image)}
    aug = clahe(**data)
    png_img = aug["image"]
    png_img = Image.fromarray(png_img)

    return image_transform(png_img), target_transform(target)
    
train_dataset = torchvision.datasets.SBDataset(root='./', image_set='train', mode='segmentation', transforms=transform)
test_dataset = torchvision.datasets.SBDataset(root='./', image_set='val', mode='segmentation', transforms=test_transform)
train_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=BATCH_S, shuffle=True)
test_loader = torch.utils.data.DataLoader(dataset=test_dataset, batch_size=BATCH_S, shuffle=False)
````

## Problem 1. Build U-Net model (30%)

- **step 1.** Build U-Net by using FCN layer (10%)
    - output channel size must be same with number of classes
    - (Hint: Use ```torch.cat``` function for concat)

- **step 2.** Train your own model with **SBD** (10%)
- **step 3.** After training, evaluate your network on test dataset and report the test accuracy. Use the pixel-wise accuracy as a metric. (10%)

$$ PixelWiseAccuracy = {NumberOfCorrectPixels \over NumberOfTotalPixels} \times 100 $$ 

- **Do not use someone's sample code or web code.** (but, you can use your own HW code)

````python
import torch
import torch.nn as nn
import torch.nn.functional as F

#TODO: define your dataset class number
CLASS_NUM = 21
def d_conv(in_channels, out_channels):
        return nn.Sequential(
        nn.Conv2d(in_channels, out_channels, 3, padding=1),
        nn.ReLU(inplace=True),
        nn.Conv2d(out_channels, out_channels, 3, padding=1),
        nn.ReLU(inplace=True)
    )

class UNet(nn.Module):

    def __init__(self, n_class=CLASS_NUM):
        super(UNet, self).__init__()
        self.dconv_down1 = d_conv(3, 64)
        self.dconv_down2 = d_conv(64, 128)
        self.dconv_down3 = d_conv(128, 256)
        self.dconv_down4 = d_conv(256, 512)
        self.dconv_down5 = d_conv(512, 1024)

        self.maxpool = nn.MaxPool2d(2)
        self.upsample4 = nn.ConvTranspose2d(1024, 1024, kernel_size=2, stride=2)
        self.upsample3 = nn.ConvTranspose2d(512, 512, kernel_size=2, stride=2)
        self.upsample2 = nn.ConvTranspose2d(256, 256, kernel_size=2, stride=2)
        self.upsample1 = nn.ConvTranspose2d(128, 128, kernel_size=2, stride=2)

        self.dconv_up4 = d_conv(512 + 1024, 512)
        self.dconv_up3 = d_conv(256 + 512, 256)
        self.dconv_up2 = d_conv(128 + 256, 128)
        self.dconv_up1 = d_conv(128 + 64, 64)
        
        self.conv_last1 = nn.Conv2d(64, n_class, 1)
   
        
    def forward(self, x):
        conv1 = self.dconv_down1(x)
        x = self.maxpool(conv1)

        conv2 = self.dconv_down2(x)
        x = self.maxpool(conv2)

        conv3 = self.dconv_down3(x)
        x = self.maxpool(conv3)

        conv4 = self.dconv_down4(x)
        x = self.maxpool(conv4)

        x = self.dconv_down5(x)
        
        x = self.upsample4(x)
        x = torch.cat([x,conv4], dim=1)

        x = self.dconv_up4(x)
        x = self.upsample3(x)
        x = torch.cat([x, conv3], dim=1) 

        x = self.dconv_up3(x) 
        x = self.upsample2(x)
        x = torch.cat([x, conv2], dim=1)

        x = self.dconv_up2(x)
        x = self.upsample1(x)
        x = torch.cat([x, conv1], dim=1)
        
        x = self.dconv_up1(x) 
        output = self.conv_last1(x) 
        

        return output
````

````python
device = torch.device('cuda:0')
num_epochs = 300
model = UNet().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(),lr=0.001)
train_loss = []
test_loss = []
train_pixacc = []
test_pixacc = []
for epoch in range(num_epochs):
    total, correct = 0, 0
    
    for i, (images, labels) in enumerate(train_loader):
        
        images, labels = images.to(device), labels.to(device)
        labels_ = labels.unsqueeze(1)


        outputs= model(images)

        loss = criterion(outputs,labels)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0) * labels.size(1) * labels.size(2)
        correct += (predicted == labels).sum().item()
        
       
    print('Epoch: {}/{}, Loss: {:.4f}, Training Accuracy: {:.4f}%'.
                format(epoch + 1, num_epochs, loss.item(), 100 * correct / total))
    
    train_pixacc.append(100 * correct / total)
    train_loss.append(loss.item())
    
    with torch.no_grad():
        total, correct = 0, 0
    
        for i, (images, labels) in enumerate(test_loader):

            images, labels = images.to(device), labels.to(device)
            labels_ = labels.unsqueeze(1)


            outputs= model(images)

            loss = criterion(outputs,labels)
           
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0) * labels.size(1) * labels.size(2)
            correct += (predicted == labels).sum().item()

            
        print('Epoch: {}/{}, Loss: {:.4f}, Testing Accuracy: {:.4f}%'.
                    format(epoch + 1, num_epochs, loss.item(), 100 * correct / total))
        
        test_pixacc.append(100 * correct / total)
        test_loss.append(loss.item())
    if epoch%20 ==0:
        torch.save(model.state_dict(),'unet_weights/std/'+str(epoch)+'.pth')
textfile = open("train_pixacc.txt", "w")
for element in train_pixacc:
    textfile.write(str(element) + "\n")
textfile.close()

textfile = open("train_loss.txt", "w")
for element in train_loss:
    textfile.write(str(element) + "\n")
textfile.close()

textfile = open("test_pixacc.txt", "w")
for element in test_pixacc:
    textfile.write(str(element) + "\n")
textfile.close()

textfile = open("test_loss.txt", "w")
for element in test_loss:
    textfile.write(str(element) + "\n")
textfile.close()

torch.save(model.state_dict(), 'standard_unet.pth')
````

````python
import torch
import torch.nn as nn
import torch.nn.functional as F

#TODO: define your dataset class number and class name
CLASS_NUM = 21
def d_conv(in_channels, out_channels):
        return nn.Sequential(
        nn.Conv2d(in_channels, out_channels, 3, padding=1),
        nn.ReLU(inplace=True),
        nn.Conv2d(out_channels, out_channels, 3, padding=1),
        nn.ReLU(inplace=True)
    )

class C_UNet(nn.Module):

    def __init__(self, n_class=CLASS_NUM):
        super(C_UNet, self).__init__()
        self.dconv_down1 = d_conv(3, 64)
        self.dconv_down2 = d_conv(64, 128)
        self.dconv_down3 = d_conv(128, 256)
        self.dconv_down4 = d_conv(256, 512)

        self.maxpool = nn.MaxPool2d(2)
        self.upsample3 = nn.ConvTranspose2d(512, 512, kernel_size=2, stride=2)
        self.upsample2 = nn.ConvTranspose2d(256, 256, kernel_size=2, stride=2)
        self.upsample1 = nn.ConvTranspose2d(128, 128, kernel_size=2, stride=2)

        self.dconv_up3 = d_conv(256 + 512, 256)
        self.dconv_up2 = d_conv(128 + 256, 128)
        self.dconv_up1 = d_conv(128 + 64, 64)
        
        self.conv_last1 = nn.Conv2d(64, n_class, 1)
   
        
    def forward(self, x):
        conv1 = self.dconv_down1(x)
        x = self.maxpool(conv1)

        conv2 = self.dconv_down2(x)
        x = self.maxpool(conv2)

        conv3 = self.dconv_down3(x)
        x = self.maxpool(conv3)

        x = self.dconv_down4(x)
        
        x = self.upsample3(x)
        x = torch.cat([x, conv3], dim=1) 

        x = self.dconv_up3(x) 
        x = self.upsample2(x)
        x = torch.cat([x, conv2], dim=1)

        x = self.dconv_up2(x)
        x = self.upsample1(x)
        x = torch.cat([x, conv1], dim=1)
        
        x = self.dconv_up1(x) 
        output = self.conv_last1(x) 
        

        return output
````

````python
device = torch.device('cuda:1')
num_epochs = 300
model = C_UNet().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(),lr=0.001)
train_loss = []
test_loss = []
train_pixacc = []
test_pixacc = []
for epoch in range(num_epochs):
    total, correct = 0, 0
    
    for i, (images, labels) in enumerate(train_loader):
        
        images, labels = images.to(device), labels.to(device)
        labels_ = labels.unsqueeze(1)


        outputs= model(images)

        loss = criterion(outputs,labels)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0) * labels.size(1) * labels.size(2)
        correct += (predicted == labels).sum().item()
        
       
    print('Epoch: {}/{}, Loss: {:.4f}, Training Accuracy: {:.4f}%'.
                format(epoch + 1, num_epochs, loss.item(), 100 * correct / total))
    
    train_pixacc.append(100 * correct / total)
    train_loss.append(loss.item())
    
    with torch.no_grad():
        total, correct = 0, 0
    
        for i, (images, labels) in enumerate(test_loader):

            images, labels = images.to(device), labels.to(device)
            labels_ = labels.unsqueeze(1)


            outputs= model(images)

            loss = criterion(outputs,labels)
           
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0) * labels.size(1) * labels.size(2)
            correct += (predicted == labels).sum().item()

            
        print('Epoch: {}/{}, Loss: {:.4f}, Testing Accuracy: {:.4f}%'.
                    format(epoch + 1, num_epochs, loss.item(), 100 * correct / total))
        
        test_pixacc.append(100 * correct / total)
        test_loss.append(loss.item())
    if epoch%20 ==0:
        torch.save(model.state_dict(),'unet_weights/custom/'+str(epoch)+'.pth')
textfile = open("c_train_pixacc.txt", "w")
for element in train_pixacc:
    textfile.write(str(element) + "\n")
textfile.close()

textfile = open("c_train_loss.txt", "w")
for element in train_loss:
    textfile.write(str(element) + "\n")
textfile.close()

textfile = open("c_test_pixacc.txt", "w")
for element in test_pixacc:
    textfile.write(str(element) + "\n")
textfile.close()

textfile = open("c_test_loss.txt", "w")
for element in test_loss:
    textfile.write(str(element) + "\n")
textfile.close()

torch.save(model.state_dict(), 'custom_unet.pth')
````
